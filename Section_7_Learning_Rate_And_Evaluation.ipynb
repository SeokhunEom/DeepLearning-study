```python
import tensorflow as tf
```

```python
# Training data: 8 samples with 3 features each
x_data = [[1, 2, 1],
          [1, 3, 2],
          [1, 3, 4],
          [1, 5, 5],
          [1, 7, 5],
          [1, 2, 5],
          [1, 6, 6],
          [1, 7, 7]]

# Labels for the training data: each sample belongs to one of three classes, represented as one-hot vectors
y_data = [[0, 0, 1],  # class 3
          [0, 0, 1],  # class 3
          [0, 0, 1],  # class 3
          [0, 1, 0],  # class 2
          [0, 1, 0],  # class 2
          [0, 1, 0],  # class 2
          [1, 0, 0],  # class 1
          [1, 0, 0]]  # class 1

# Test data: 3 samples with 3 features each
x_test = [[2, 1, 1],
          [3, 1, 2],
          [3, 3, 4]]

# Labels for the test data: each sample belongs to one of three classes, represented as one-hot vectors
y_test = [[0, 0, 1],  # class 3
          [0, 0, 1],  # class 3
          [0, 0, 1]]  # class 3
```

```python
# Set the learning rate for the optimizer
learning_rate = 0.1

# Initialize a Sequential model
tf.model = tf.keras.Sequential()
```

```python
# Add a Dense layer with 3 units (one for each class), input dimension of 3 (number of features), and softmax activation function
# Softmax is used to output a probability distribution over the classes
tf.model.add(tf.keras.layers.Dense(units=3, input_dim=3, activation='softmax'))
```

```python
# Compile the model with categorical crossentropy loss, SGD optimizer with the specified learning rate, and accuracy as a metric
tf.model.compile(loss='categorical_crossentropy',
                 optimizer=tf.keras.optimizers.SGD(learning_rate=learning_rate),
                 metrics=['accuracy'])
```

```python
# Train the model using the training data (x_data, y_data) for 1000 epochs
tf.model.fit(x_data, y_data, epochs=1000)
```

```python
# Make predictions on the test data
# Note: In TensorFlow 2.x, the predict_classes method has been deprecated. Use predict and then argmax instead.
# Convert predicted probabilities to class labels
predictions = tf.model.predict(x_test)
predicted_classes = tf.argmax(predictions, axis=1)
print("Prediction: ", predicted_classes.numpy())
```

```python
# Evaluate the model on the test data and print the accuracy
# evaluate method returns a list [loss, accuracy]
accuracy = tf.model.evaluate(x_test, y_test)[1]
print("Accuracy: ", accuracy)
```
